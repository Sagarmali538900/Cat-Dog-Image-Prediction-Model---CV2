```python
from tensorflow.keras.models import load_model
```

```python
model=load_model('mymodel.h5')
```

```python
import cv2
```

```python
def get_class(path):
    A=cv2.imread(path)
    A=cv2.resize(A,(224,224))
    A=A.reshape(1,224,224,3)/255
    print(model.predict(A))
    if model.predict(A).argmax()==0:
        return('cat')
    else:
        return('dog')
        
    
    
```

```python
get_class('')
```
